# Binomial with Beta Prior

## Motivating Study: The Magician's Coin

> One day while walking downtown, you stop briefly to watch a street-performer - a magician - who is doing various sleight-of-hand tricks for a crowd. As a part of his next trick, he shows the crowd a standard-appearing coin with both head and tails sides. Somewhat abruptly the magician singles you out from the crowd and offers you a deal: if he flips the coin 10 times and it lands on heads 7 or less times, he will pay you 100 dollars, but if it lands on heads 8 or more times, you owe him 100 dollars. The question: do you take the bet?

## Calculating the Expected Value of A Fair Coin

Let us first calculate the expected value of the above proposition assuming the coin is fair.

To begin we model the number of heads as a random variable over $n$ coin flips with $s$ number of heads and $n-s$ tails according to some probability $p$ of landing on heads and a complementary probability $(1-p)$ of landing on tails. 

|Parameter|Interpretation|
|:-:|:-:|
|$n\in\mathbb{N}$|number of trials|
|$p\in[0,1]$|probability of trial success |
|$\{s:s\in\mathbb{N}_0\wedge s\le n\}$|number of successes in $n$ trials|

With this parameterization, the random variable will follow the binomial distribution with a probability mass function (PMF) of the form 

$$P(s|n,p)={n \choose s}p^{s}(1-p)^{n-s}$$

We begin by graphing the probability of each combination of heads and tails and coloring it by its associated betting outcome, in addition to calculating the expected value of the bet.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# set parameter values
n = 10 # number of flips
p = 0.5 # probability of heads

#define possible outcomes of heads
x_values = np.arange(n+1)

#calculate pmf values
dist = np.array([binom.pmf(x,n,p) for x in x_values])

#create color masks
win = x_values<=7
lose = x_values>7

#plot the distribution
plt.bar(x_values[win],dist[win],color='green',label='Win $100')
plt.bar(x_values[lose],dist[lose],color='red',label='Lose $100')
plt.ylabel('Probability',fontsize='15')
plt.xlabel('Number of Heads',fontsize='15')
plt.legend()
plt.show()

# calculate expected value
expected_value = 100*sum(dist[win]) - 100*sum(dist[lose])
print("Expected Value:",f"${round(expected_value,2)}")
```

Given the above calculation, the proposition sounds like a wonderful deal with an expected value of $89.06 in your favor; however, how do you know whether the coin is fair? Better yet, how would you quantify your uncertainty about the fairness of the coin? Suppose, in this hypothetical (although unlikely in reality) that the magician allows you to flip the coin a few times prior to taking the bet. How could you use the data that you collect to inform your decision by updating your uncertainty?

## Parameterizing a Possibly Unfair Coin

The above parameterization of the coin assumes that the coin is fair (i.e. $p=0.5$). Instead, let us instead allow $p$ to be random variable with its own probability distribution. To model $P(p)$, we will use the the beta distribution due to its conjugacy to the binomial distribution.

$$P(p)=Beta(p|\alpha,\beta)=\frac{p^{\alpha-1}(1-p)^{\beta-1}}{B(\alpha,\beta)}$$

where $\alpha$ and $\beta$ are hyperparameters encoding prior belief or information and $B(\alpha,\beta)$ represents the Beta function serving as a normalizing constant.

Applying [Bayes' theorem](definitions.ipynb#bayes) to the binomial likelihood $P(s|n,p)$ with the beta distribution as the prior $P(p)$, we derive the posterior distribution of the binomial distribution $P(p|n,s)$.

$$
\begin{align}
P(p|n,s) & = \frac{P(n,s|p)P(p)}{\int_p P(n,s|p)P(p)dp} \\
& = \frac{{n \choose s}p^{s+\alpha-1}(1-p)^{n-s+\beta-1}/B(\alpha,\beta)}{\int_{0}^{1} {n \choose s}p^{s+\alpha-1}(1-p)^{n-s+\beta-1}/B(\alpha,\beta)dp} \\
& = \frac{p^{s+\alpha-1}(1-p)^{n-s+\beta-1}}{B(s+\alpha,n-s+\beta)} \\
& = Beta(p|\alpha'=s+\alpha,\beta'=n-s+\beta) \\
\end{align}
$$

which is a Beta distribution with parameters $(\alpha'=s+\alpha,\beta'=n-s+\beta)$, proving the conjugacy of the binomial and beta distributions.

Thus, the hyperparameters of the posterior distribution can be found analytically using only the following relation:

$$P(p|\alpha',\beta')=Beta(p|\alpha',\beta')=Beta(p|\alpha+s,\beta+n-s)$$

### Intuition of the Beta Distribution Hyperparameters

The $\alpha$ and $\beta$ hyperparameters of the Beta prior distribution can seem somewhat cryptic at first, until one recognizes that they represent pseudo-observations of the prior. A Beta prior with hyperparameters $(1,1)$ is used as an uninformative prior as it represents a coin where one has witnessed no coin flips. Meanwhile a Beta prior of $(86,16)$ can be understood as a coin which we have flipped 100 times, with 85 heads and 15 tails. Thus in the process of updating our Beta prior using [Bayes' theorem](appendixA.ipynb#Bayes'-Theorem), it makes sense that the resulting posterior distribution would have parameters $(\alpha'=\alpha+s,\beta'=\beta+n-s)$ as it simply represents the addition of real observations of the data to the pseudo-observations of the prior.



## Estimating the Bias of a Coin

Revisiting the motivating study, suppose you test the coin by flipping it 50 times and it lands on heads exactly 42 times and tails 8 times. Assuming an uninformative prior of $Beta(1,1)$, the posterior would have the form of $Beta(1+42,1+8)=Beta(43,9)$.

```python
from scipy.stats import beta as beta_distribution

def plot_beta(alpha,beta,color):
    theta = np.linspace(0,1,100)
    dist = beta_distribution.pdf(theta,alpha,beta)
    plt.plot(theta,dist,color,label=f"Beta({alpha},{beta})")

# set parameter values
prior_alpha, prior_beta = 1,1
posterior_alpha, posterior_beta = 43, 9

# plot the prior and posterior distributions
plot_beta(prior_alpha, prior_beta,color="red")
plot_beta(posterior_alpha, posterior_beta,color="blue")
plt.xlabel('p (probability of heads)',fontsize='15')
plt.ylabel('Probability',fontsize='15')
plt.legend()
plt.show()
```

### Methods of Describing the Posterior Distribution

As one can see in the above plot of the uninformative prior and posterior distributions, the uncertainty in the model parameter $p$ after the observed data is incorporated into the model, manifesting as a concentration of the probability density around the most probable value of $p$.

In describing the binomial posterior distribution, there are two main values of interest:
1. the [maximum a posteriori  (MAP)](appendixA.ipynb#MAP), which represents the most probable value of the $p$ given the data
2. the [highest density interval (HDI)](appendixA.ipynb#Highest-Density-Interval-(HDI)), which is the narrowest interval over the posterior distribution in which the parameter $p$ falls with a given total probability (i.e. a 95% credible interval)

```python
from scipy import optimize

#define function to calculate highest density interval
def hdi_unimodal(distribution, level=0.95):
    """
    Get the highest density interval for the distribution, e.g. for a Bayesian posterior, the highest posterior density interval (HPD/HDI)
    Credit: Ying Tong Li (https://yingtongli.me/blog/2022/10/06/scipy-hdi.html) 
    """

    # For a given lower limit, we can compute the corresponding 95% interval
    def interval_width(lower):
        upper = distribution.ppf(distribution.cdf(lower) + level)
        return upper - lower

    # Find such interval which has the smallest width
    # Use equal-tailed interval as initial guess
    initial_guess = distribution.ppf((1-level)/2)
    optimize_result = optimize.minimize(interval_width, initial_guess,method = 'Nelder-Mead')

    lower_limit = optimize_result.x[0]
    width = optimize_result.fun
    upper_limit = lower_limit + width

    return (lower_limit, upper_limit)

# update plot_beta to include MAP and HDI
def plot_beta(alpha,beta,color,annotations=False,level=.95):
    theta = np.linspace(0,1,100)
    dist = beta_distribution.pdf(theta,alpha,beta)
    plt.plot(theta,dist,color,label=f"Beta({alpha},{beta})")
    if annotations:
        MAP_theta = max(dist)
        MAP_p = theta[np.argmax(dist)]
        hdi_lower, hdi_upper = hdi_unimodal(beta_distribution(alpha,beta),level)
        plt.vlines(x=MAP_p,ymin=0,ymax=MAP_theta,color=color,label=f"MAP: {round(MAP_p,2)}")
        plt.fill_between(theta,y1=dist,where=(hdi_lower<theta)&(theta<hdi_upper),color=color,alpha=0.2,label=f"{round(level*100)}% HDI: [{round(hdi_lower,2)},{round(hdi_upper,2)}]")
    

plot_beta(prior_alpha, prior_beta,color="red")
plot_beta(posterior_alpha, posterior_beta,color="blue",annotations=True)
plt.xlabel('p (probability of trial success)',fontsize='15')
plt.ylabel('Probability of p given data',fontsize='15')
plt.legend()
plt.show()
```

## Calculating the Expected Value of a Biased Coin

The expected value of a binomial-distributed variable, such as the biased coin of this example, can be calculated using its [posterior predictive distribution (PPD)](definitions.ipynb#PPD) $P(s|n,\alpha',\beta')$, which can be found by marginalizing the binomial likelihood $P(s|n,p)$ over the beta-distributed posterior of $P(p|\alpha',\beta')$.


$$\begin{align}
PPD &= P(s|n,\alpha',\beta') \\
&= \int_{p}P(s|n,p)P(p|\alpha',\beta')dp \\
& = \int_0^1Bin(s|n,p)Beta(p|\alpha',\beta')dp \\
& = {n \choose s}\frac{1}{B(\alpha',\beta')}\int_0^{1}p^{s+\alpha'-1}(1-p)^{n-s+\beta'-1}dp \\
& = {n \choose s}\frac{B(s+\alpha',n-s+\beta')}{B(\alpha',\beta')} \\
& = BetaBin(n,\alpha',\beta') \\
\end{align}
$$

Plotting the beta-binomial distribution and coloring it by its associated betting outcome, we are able to calculate the expected value of the bet.

```python
from scipy.stats import betabinom

# set parameter values
n = 10 # number of flips
alpha = 43 # number of observed successes + 1
beta = 9 # number of observed failures + 1

#define possible outcomes of success/heads
x_values = np.arange(n+1)

#calculate pmf values
dist = np.array([betabinom.pmf(x,n,alpha,beta) for x in x_values])

#create color masks
win = x_values<=7
lose = x_values>7

#plot the distribution
plt.bar(x_values[win],dist[win],color='green',label='Win $100')
plt.bar(x_values[lose],dist[lose],color='red',label='Lose $100')
plt.ylabel('Probability',fontsize='15')
plt.xlabel('Number of Heads',fontsize='15')
plt.legend()
plt.show()

# calculate expected value of the bet
expected_value = 100*sum(dist[win]) - 100*sum(dist[lose])
print("Expected Value:",f"${round(expected_value,2)}")
```

Thus, we would could expect to lose an average of $49.40 every time we make the bet with the magician, given the observed data of 42 heads in 50 tosses.
